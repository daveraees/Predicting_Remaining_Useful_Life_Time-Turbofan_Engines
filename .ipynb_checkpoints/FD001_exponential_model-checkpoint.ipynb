# loading the datasets

```python
trainData = []
testData = []
RUL = []
#names = ['unit', 'time']
names = ['unit', 'time']
names.extend([('op%d' %  j ) for j in range(1,4)])
names.extend([('sr%1d' % j) for j in range(1,22)])

for i in range(1,5):
    trainData.append(pd.read_table(('train_FD00%d.txt' % i),
                                  sep=' ',
                                  index_col=False,
                                  names=names,
                                  header = None,
                                  decimal='.',
                                  engine='python'))
    trainData[-1].set_index(['unit', 'time'],inplace=True)
    testData.append(pd.read_table(('test_FD00%d.txt' % i),
                                  sep=' ',
                                  index_col=False,
                                  names=names,
                                  header = None,
                                  decimal='.',
                                  engine='python'))
    testData[-1].set_index(['unit', 'time'],inplace=True)
    RUL.append(pd.read_table(('RUL_FD00%d.txt' % i),
                             sep=' ',
                             index_col=False,
                             names=['RULtime'],
                             header = None,
                             engine='python'))

    
```

## exponential model

```python



# Function to calculate the exponential with constants a and b
def exponential(x, A, k, y):
    return (A*np.exp(-k*(x))  + y)
            
    
base_span = 30

fit_expDF = pd.DataFrame(index=np.arange(1,FD001["Train trjectories:"]),
                             columns=['A', 'k', 'y0', 'Astd', 'kstd', 'ystd'])
fit_expDF.rename_axis(['unit',], inplace=True,axis=0)

nrows=10
ncols=10
fig, ax = pp.subplots(figsize=(30,20),nrows=2*nrows, ncols=ncols, sharex=True, sharey=True,)
for traj_jdx in range (0,nrows,1):
    for traj_idx in range(0,ncols,1): 
        #operational parameters on every even row
        row=traj_jdx*2
        col=traj_idx
        data_idx = ncols*traj_jdx+traj_idx+1
        ax[row,col].plot(train_dataPD.loc[data_idx,'op1']*1000)
        ax[row,col].plot(train_dataPD.loc[data_idx,'op2']*5000,alpha=0.4)
        ax[row,col].axhline(y=0)
        # PC of the sensor readings on the every odd row
        trace=train_dataPD.loc[data_idx,'pc1']
        ax[row+1,col].plot(trace)
        
        base = trace.iloc[-base_span:].mean()
        ax[row+1,col].axhline(y=base)
        
        X = trace.index.values
        y = trace.values.astype(np.float)
        
        # Fit the experimental data
        pars, cov = curve_fit(f=exponential, xdata=X, ydata=y, p0=[8-base, 0.02, -0.1], bounds=(-np.inf, np.inf))
        perr = np.sqrt(np.diag(cov)) # standard deviation errors on the parameters
        pars_errs = np.hstack([pars,perr])
        
        #print('score: ',reg.score(X, y), )
        yy = exponential(X, *pars)
        ax[row+1,col].plot(X,yy)
        fit_expDF.loc[data_idx,:] = pars_errs
        
        

```

The exponential model fits the data well. I will try to use it for prediction.

### try fit exponential to the test data and predict the RUL based on the fit only

#### the time reversal for the test data

```python
# reverse the order of time in each trajectory
for traject in range(1,FD001['Test trajectories:']+1):
    traj = test_dataPD.loc[traject,:].values
    test_dataPD.loc[traject,:] = np.flipud(traj)
```

```python
# rename the index
test_dataPD.rename_axis(['unit','rtime'], inplace=True,axis=0)
```

```python
test_dataPD.head()
```

```python
base_span = 20


# Function to calculate the exponential with constants a and b
def exponential(x, k, y, RUL):
    # the x means the reverse time, breakdown is expected at x-RUL = 0
    threshold = pc1_target # calculated mean value of the terninal PC1 signal
    A=threshold-y
    return (A*np.exp(-k*(x+RUL))  + y)


fit_expDF = pd.DataFrame(index=np.arange(1,FD001['Test trajectories:']),
                             columns=[ 'k', 'y0', 'RUL', 'kstd', 'ystd', 'RULstd'])
fit_expDF.rename_axis(['unit',], inplace=True,axis=0)

nrows=10
ncols=10
fig, ax = pp.subplots(figsize=(30,20),nrows=2*nrows, ncols=ncols, sharex=True, sharey=True,)
for traj_jdx in range (0,nrows,1):
    for traj_idx in range(0,ncols,1): 
        #operational parameters on every even row
        row=traj_jdx*2
        col=traj_idx
        data_idx = ncols*traj_jdx+traj_idx+1
        ax[row,col].plot(test_dataPD.loc[data_idx,'op1']*1000)
        ax[row,col].plot(test_dataPD.loc[data_idx,'op2']*5000,alpha=0.4)
        ax[row,col].axhline(y=0)
        # PC of the sensor readings on the every odd row
        trace=test_dataPD.loc[data_idx,'pc1']
        ax[row+1,col].plot(trace)
        
        base = trace.iloc[-base_span:].mean()
        ax[row+1,col].axhline(y=base)
        
        X = trace.index.values
        y = trace.values.astype(np.float)
        
        # Fit the experimental data
        p0 = [0.02, base,+100]
        try:
            pars, cov = curve_fit(f=exponential, xdata=X, ydata=y, p0=p0, bounds=(-np.inf, np.inf),check_finite=False)
            perr = np.sqrt(np.diag(cov)) # standard deviation errors on the parameters
            pars_errs = np.hstack([pars,perr])
        except:
            print('test trajectory %d could not be fitted with the given start parameters' % data_idx)
            pars_errs = np.hstack([p0,[0,0,0]])
        yy = exponential(X, *pars)
        ax[row+1,col].plot(X,yy)
        fit_expDF.loc[data_idx,:] = pars_errs
        
        
```
